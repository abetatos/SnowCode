```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

PATH = "/home/abetatos/Escritorio/TFM_src/train/Models/"
files = os.listdir(PATH)
files = [f for f in files if f.endswith("train_report.csv")]
vals = []
for file in files: 
    df = pd.read_csv(PATH + file)
    df2 = pd.read_csv(PATH + file.replace("train_report", "best_report"))
    val = df2['Train loss'].iloc[0]
    feature = json.loads(df2['Features'].iloc[0].replace("'", '"'))[0]
    vals.append([val, feature])
    plt.ylim((0.5, 1))
    # print(file)
    plt.plot(df['Val loss 2'].values, label=feature)
    # plt.show()
plt.legend()
```
